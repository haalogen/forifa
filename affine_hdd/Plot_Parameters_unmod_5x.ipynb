```python
import numpy as np
import datetime as dt

%pylab inline
```

```python
date = ['2016-02-18', '2016-02-19',
        '2016-03-23', '2016-03-29', '2016-05-07']

# a_dis = [0.998, 0.993, 0.995, 0.996, 0.997]
a_aff = [1.0085, 0.997,  0.995, 1.0029, 0.991]

# b_dis = [  0.011, 0.009,  0.01,  0.014, 0.008]
b_aff = [0.019, 0.0005, -0.0246, 0.00121, -0.00295]

# c_dis = [-0.011, -0.011, -0.013, -0.015, -0.011, -0.013]
c_aff = [-0.0149, -0.0068, -0.0171, 0.00252, -0.0115]

# d_dis = [0.997, 0.989, 0.999, 0.997, 1.000, 0.999]
d_aff = [1.007, 0.993, 1.0043, 0.990, 0.976]

# e_dis = [236, 231, 240, 235, 227, 240]
e_aff = [229, 250, 254, 247, 254]

# f_dis = [63, 66, 64, 58, 52, 55]
f_aff = [52, 54, 47, 45, 54]

# eps1 = [3.01, 3.01, 2.78, 2.94, 3.43, 2.91]
# eps2 = [3.12, 3.06, 2.93, 3.00, 3.39, 2.93]
 
x = np.arange(5)
```

```python
# pylab.plot(x, a_dis, '--ro', label='a_dist')
pylab.plot(x, a_aff, '-.b^', label='a_aff')
pylab.xticks(x, date, rotation=70)

pylab.legend(loc='upper right')
pylab.title('Coefficient A')
# xmin, xmax = pylab.xlim()
# ymin, ymax = pylab.ylim()
```
