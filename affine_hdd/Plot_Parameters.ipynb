```python
import numpy as np
import datetime as dt

%pylab inline
```

```python
date = ['2015-10-16', '2016-02-18', '2016-02-19',
        '2016-03-23', '2016-03-29', '2016-05-07']

a_dis = [0.998, 0.998, 0.993, 0.995, 0.996, 0.997]
a_aff = [1.01,  0.997, 1.01,  0.996, 1.003, 1.005]

b_dis = [0.006,  0.011, 0.009,  0.01,  0.014, 0.008]
b_aff = [-0.002, 0.026, 0.005, -0.002, 0.009, -0.007]

c_dis = [-0.011, -0.011, -0.013, -0.015, -0.011, -0.013]
c_aff = [-0.013, -0.015, -0.018, -0.023, -0.015, -0.020]

d_dis = [0.997, 0.989, 0.999, 0.997, 1.000, 0.999]
d_aff = [0.997, 0.980, 1.010, 0.998, 1.004, 1.001]

e_dis = [236, 231, 240, 235, 227, 240]
e_aff = [246, 239, 234, 261, 242, 262]

f_dis = [63, 66, 64, 58, 52, 55]
f_aff = [69, 80, 66, 69, 55, 61]

eps1 = [3.01, 3.01, 2.78, 2.94, 3.43, 2.91]
eps2 = [3.12, 3.06, 2.93, 3.00, 3.39, 2.93]
 
x = np.arange(6)
```

```python
pylab.plot(x, a_dis, '--ro', label='a_dist')
pylab.plot(x, a_aff, '-.b^', label='a_aff')
pylab.xticks(x, date, rotation=70)

pylab.legend(loc='upper right')
pylab.title('Coefficient A')
# xmin, xmax = pylab.xlim()
# ymin, ymax = pylab.ylim()
```

```python
pylab.plot(x, b_dis, '--ro', label='b_dist')
pylab.plot(x, b_aff, '-.b^', label='b_aff')
pylab.xticks(x, date, rotation=70)

pylab.legend(loc='upper right')
pylab.title('Coefficient B')
```

```python
pylab.plot(x, c_dis, '--ro', label='c_dist')
pylab.plot(x, c_aff, '-.b^', label='c_aff')
pylab.xticks(x, date, rotation=70)

pylab.legend(loc='lower left')
pylab.title('Coefficient C')
```

```python
pylab.plot(x, d_dis, '--ro', label='d_dist')
pylab.plot(x, d_aff, '-.b^', label='d_aff')
pylab.xticks(x, date, rotation=70)

pylab.legend(loc='lower right')
pylab.title('Coefficient D')
```

```python
pylab.plot(x, e_dis, '--ro', label='e_dist')
pylab.plot(x, e_aff, '-.b^', label='e_aff')
pylab.xticks(x, date, rotation=70)

pylab.legend(loc='upper left')
pylab.title('Coefficient E')
```

```python
pylab.plot(x, f_dis, '--ro', label='f_dist')
pylab.plot(x, f_aff, '-.b^', label='f_aff')
pylab.xticks(x, date, rotation=70)

pylab.legend(loc='upper right')
pylab.title('Coefficient F')
```

```python
pylab.plot(x, eps1, '--ro', label='eps1')
pylab.plot(x, eps2, '-.b^', label='eps2')
pylab.xticks(x, date, rotation=70)

pylab.legend(loc='upper left')
pylab.title('Distortion coefficients EPS1, EPS2')
```

```python
print 'A_aff mean:%.4f, std:%.4f' % (np.mean(a_aff), np.std(a_aff))
print 'A_dis mean:%.4f, std:%.4f' % (np.mean(a_dis), np.std(a_dis))
```

```python
print 'B_aff mean:%.4f, std:%.4f' % (np.mean(b_aff), np.std(b_aff))
print 'B_dis mean:%.4f, std:%.4f' % (np.mean(b_dis), np.std(b_dis))
```

```python
print 'C_aff mean:%.4f, std:%.4f' % (np.mean(c_aff), np.std(c_aff))
print 'C_dis mean:%.4f, std:%.4f' % (np.mean(c_dis), np.std(c_dis))
```

```python
print 'D_aff mean:%.4f, std:%.4f' % (np.mean(d_aff), np.std(d_aff))
print 'D_dis mean:%.4f, std:%.4f' % (np.mean(d_dis), np.std(d_dis))
```

```python
print 'E_aff mean:%.4f, std:%.4f' % (np.mean(e_aff), np.std(e_aff))
print 'E_dis mean:%.4f, std:%.4f' % (np.mean(e_dis), np.std(e_dis))
```

```python
print 'F_aff mean:%.4f, std:%.4f' % (np.mean(f_aff), np.std(f_aff))
print 'F_dis mean:%.4f, std:%.4f' % (np.mean(f_dis), np.std(f_dis))
```
